```python
import matplotlib
import matplotlib.pyplot as plt
from numpy import arange
from numpy import meshgrid

matplotlib.use("pgf")
matplotlib.rcParams.update({
    "pgf.texsystem": "pdflatex",
    'font.family': 'serif',
    'text.usetex': True,
    'pgf.rcfonts': False,
})

%matplotlib inline

a1 = 0.02
w1 = 0.3
a2 = 0.02
w2 = 0.5
eps = 0.001
n = 800

delta = 0.025
xrange = arange(0,1,delta)
yrange = arange(0,1,delta)
X, Y = meshgrid(xrange, yrange)

F = (1 - X) * (a1 + (eps * n * (X**2 + Y**2))) - w1 * X
G = (1 - Y) * (a2 + (2 * eps * n * X * Y)) - w2 * Y

plot1=plt.contour(X, Y, F, [0], colors='#930000').collections[0]
plot2=plt.contour(X, Y, G, [0], colors='#009307').collections[0]
plt.xlabel('$p^*$', fontsize='large')
plt.ylabel('$q^*$', fontsize='large', rotation='horizontal')

plot1.set_label(r'$(1 - p^*)(\alpha + \epsilon n_1((p^*)^2+(q^*)^2) - \omega p^* = 0$')
plot2.set_label(r'$(1 - q^*)(\tilde{\alpha} + 2 \epsilon n_1 p^* q^*) - \tilde{\omega} q^*=0$')

plt.legend(loc='upper left', labelcolor=['#930000', '#009307'], frameon=False)
plt.savefig('plot1.pgf')

plt.show()
```
